```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

# https://medium.com/@olutosinbanjo/how-to-plot-a-direction-field-with-python-1fd022e2d8f8
```

```python
# parameters related to the model
R = 2
Lambda = 1/5
Delta = 1/6 # 1
```

```python
nx, ny = 0.04, 0.04
x = np.arange(0, 1+nx, nx)
y = np.arange(0, 1+nx, ny)
# print(x)
# print(y)
```

```python
X, Y  = np.meshgrid(x, y)
# print(X)
# print(Y)

print(X[0][0])
print(Y[0][0])

print(X[0][1])
print(Y[0][1])

print(X[1][0])
print(Y[1][0])
```

```python
dy = 1 - Y - (Y/Lambda) * np.abs(-Y + R*X)
dx = Delta * (1 - X) - (X/Lambda) * np.abs(-Y + R*X)
```

```python
plt.figure(dpi = 150)
plt.quiver(X, Y, dx, dy, color = 'purple')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
# plt.figure(dpi = 200)
# plt.quiver(X, Y, dx, dy, color = 'purple', headaxislength = 0, headlength = 0)
# plt.show()         
```

```python
dyu = dy/np.sqrt(dx**2 + dy**2)
dxu = dx/np.sqrt(dx**2 + dy**2)
```

```python
plt.figure(dpi = 200)
plt.quiver(X, Y, dxu, dyu, color = 'purple')
# ax = plt.gca()
# ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
# now zoom on the fixed points

# fixed point 1: stable node
nx, ny = 0.003, 0.003
x_min = 0.1
x_max = 0.16
y_min = 0.45
y_max = 0.52
x = np.arange(x_min, x_max+nx, nx)
y = np.arange(y_min, y_max+nx, ny)
x_fixed_point = 0.13499
y_fixed_point = 0.48356

X, Y  = np.meshgrid(x, y)

dy = 1 - Y - (Y/Lambda) * np.abs(-Y + R*X)
dx = Delta * (1 - X) - (X/Lambda) * np.abs(-Y + R*X)

# plt.figure(dpi = 200)
# plt.quiver(X, Y, dx, dy, color = 'purple')
# plt.scatter(x_fixed_point, y_fixed_point, marker = 'o', c = 'r', s = 2)
# plt.show()

dyu = dy/np.sqrt(dx**2 + dy**2)
dxu = dx/np.sqrt(dx**2 + dy**2)

plt.figure(dpi = 200)
plt.quiver(X, Y, dxu, dyu, color = 'purple')
plt.scatter(x_fixed_point, y_fixed_point, marker = 'o', c = 'r', s = 5)
plt.show()
```

```python
# fixed point 3: stable spiral
nx, ny = 0.0005, 0.001
x_min = 0.43
x_max = 0.44
y_min = 0.81
y_max = 0.84
x = np.arange(x_min, x_max+nx, nx)
y = np.arange(y_min, y_max+nx, ny)
x_fixed_point = 0.43215
y_fixed_point = 0.82034

X, Y  = np.meshgrid(x, y)

dy = 1 - Y - (Y/Lambda) * np.abs(-Y + R*X)
dx = Delta * (1 - X) - (X/Lambda) * np.abs(-Y + R*X)

plt.figure(dpi = 200)
plt.quiver(X, Y, dx, dy, color = 'purple')
plt.scatter(x_fixed_point, y_fixed_point, marker = 'o', c = 'r', s = 2)
plt.show()

dyu = dy/np.sqrt(dx**2 + dy**2)
dxu = dx/np.sqrt(dx**2 + dy**2)

plt.figure(dpi = 200)
plt.quiver(X, Y, dxu, dyu, color = 'purple')
plt.scatter(x_fixed_point, y_fixed_point, marker = 'o', c = 'r', s = 5)
plt.show()
```
